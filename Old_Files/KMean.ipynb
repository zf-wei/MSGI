```python
%%time
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(0)
X = np.concatenate([
    np.random.randn(100, 2) * 0.5 + np.array([2, 3]),
    np.random.randn(100, 2) * 0.5 + np.array([-2, -2]),
    np.random.randn(100, 2) * 0.5 + np.array([-2, 2])
])


# Plot the data
plt.scatter(X[:, 0], X[:, 1], color='b')
plt.title("2-Dimensional Sample")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
%%time
# Use SKlearn package and its method for KMeans
from sklearn.cluster import KMeans
# Create a K-means clustering model
kmeans = KMeans(n_clusters=3, n_init=10)

# Fit the model to the data
kmeans.fit(X)

# Get the cluster labels for each data point
labels = kmeans.labels_

# Get the cluster centers
centers = kmeans.cluster_centers_

cmap = matplotlib.colormaps.get_cmap('viridis') #This is in fact the deault color option

# Plot the data points with color-coded clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap=cmap)
plt.scatter(centers[:, 0], centers[:, 1], marker='*', c='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('K-means Clustering')
#plt.colorbar(label='Cluster')
plt.show()
```
