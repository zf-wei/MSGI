```python
# Necessary Packages
import networkx as nx

from networkx.generators.community import LFR_benchmark_graph

import matplotlib
import matplotlib.pyplot as plt

from pyspark.ml.clustering import KMeans

from pyspark.sql import SparkSession
from pyspark.ml.linalg import Vectors


from sklearn.metrics import normalized_mutual_info_score
from clusim.clustering import Clustering
import clusim.sim as sim
```

```python
# Generate LFR benchmark graph
n = 1000
tau1 = 2  # Power-law exponent for the degree distribution
tau2 = 1.1 # Power-law exponent for the community size distribution 
            #S hould be >1
mu = 0.1 # Mixing parameter

G = LFR_benchmark_graph(
    n, tau1, tau2, mu, average_degree=25, max_degree=100, min_community=50, max_community=100, 
    seed=2
)
                        
G = nx.Graph(G) # Remove multi-edges

selfloop_edges = list(nx.selfloop_edges(G)) # a list of self loops

G.remove_edges_from(selfloop_edges) # Remove self-loops
```

```python
# Get the intrinsic community structure as a list
intrinsic_communities = {frozenset(G.nodes[v]["community"]) for v in G}
intrinsic_membership = {}
for node in range(G.number_of_nodes()):
    for index, inner_set in enumerate(intrinsic_communities):
        if node in inner_set:
            intrinsic_membership[node] = index
            break
print(f"The network has {max(intrinsic_membership.values())+1} intrinsic communities.")

intrinsic_clustering = Clustering(elm2clu_dict={i: [intrinsic_membership[i]] for i in intrinsic_membership.keys()})
```

```python
import networkx as nx
import matplotlib.pyplot as plt


# 获取度序列
degree_sequence = [degree for _, degree in G.degree()]


# 绘制度序列的分布曲线
plt.hist(degree_sequence, bins='auto', density=True, alpha=0.7)
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.title('Degree Distribution')
plt.grid(True)
plt.show()
```

```python
import networkx as nx
from sklearn.manifold import LocallyLinearEmbedding
import matplotlib.pyplot as plt
import numpy as np
from networkx.generators.community import LFR_benchmark_graph

for N in range(20,50):
    
    # Perform Locally Linear Embedding
    lle = LocallyLinearEmbedding(n_neighbors=N, n_components=2)
    X_embedded = lle.fit_transform(nx.to_numpy_array(G))

    # Extract community membership
    communities = [frozenset(c) for c in nx.algorithms.community.greedy_modularity_communities(G)]
    community_colors = np.zeros(n)
    for i, community in enumerate(communities):
        community_colors[list(community)] = i

    # Plot the embedded graph with community colors and edges
    fig, ax = plt.subplots()
    #for edge in G.edges():
    #    ax.plot([X_embedded[edge[0], 0], X_embedded[edge[1], 0]], [X_embedded[edge[0], 1], X_embedded[edge[1], 1]], 'k-', alpha=0.01)
    scatter = ax.scatter(X_embedded[:, 0], X_embedded[:, 1], c=community_colors, cmap=plt.cm.get_cmap('tab20', len(communities)))
    ax.set_title(f"Locally Linear Embedding with Community Colors, n_neighbor = {N}")
    plt.colorbar(scatter, ticks=range(len(communities)))
    plt.show()
```

```python
import networkx as nx
from sklearn.manifold import LocallyLinearEmbedding
import matplotlib.pyplot as plt
import numpy as np
from networkx.generators.community import LFR_benchmark_graph

K = 14
for N in range(3,20):
    lle = LocallyLinearEmbedding(n_neighbors=25, n_components=N)
    X_embedded = lle.fit_transform(nx.to_numpy_array(G))

    lle_vec = [Vectors.dense(row) for row in X_embedded]
    lle_prep = SparkSession.builder.getOrCreate().createDataFrame([(vector,) for vector in lle_vec], ["embd"])

    # from pyspark.ml.clustering import KMeans

    # Create and fit the KMeans model
    euclid_kmeans = KMeans(k=K, featuresCol="embd")
    cosine_kmeans = KMeans(k=K, featuresCol="embd", distanceMeasure="cosine")
    lle_euclid_model = euclid_kmeans.fit(lle_prep)
    lle_cosine_model = cosine_kmeans.fit(lle_prep)

    # Add the cluster assignment to the DataFrame
    lle_euclid = lle_euclid_model.transform(lle_prep)
    lle_cosine = lle_cosine_model.transform(lle_prep)

    # hope_euclid.show(truncate=False) # 这条语句可以查看

    # Extract the cluster assignment and convert it to a list
    lle_euclid_membership = lle_euclid.select("prediction").rdd.flatMap(lambda x: x).collect()
    lle_cosine_membership = lle_cosine.select("prediction").rdd.flatMap(lambda x: x).collect()

    # 计算
    # 内在的社群结构 intrinsic_membership 和
    # 嵌入聚类算法
    # 的 NMI 以及 Element centric similarity
    # from sklearn.metrics import normalized_mutual_info_score

    # Calculate the NMI
    lle_euclid_nmi = normalized_mutual_info_score(list(lle_euclid_membership), list(intrinsic_membership.values()))
    lle_cosine_nmi = normalized_mutual_info_score(list(lle_cosine_membership), list(intrinsic_membership.values()))

    print(f"{N}-NMIs:", lle_euclid_nmi, lle_cosine_nmi)

    lle_euclid_clustering = Clustering(elm2clu_dict={i: [lle_euclid_membership[i]] for i in range(len(lle_euclid_membership))})
    lle_cosine_clustering = Clustering(elm2clu_dict={i: [lle_cosine_membership[i]] for i in range(len(lle_cosine_membership))})

    lle_euclid_similarity = sim.element_sim(intrinsic_clustering, lle_euclid_clustering, alpha=0.9)
    lle_cosine_similarity = sim.element_sim(intrinsic_clustering, lle_cosine_clustering, alpha=0.9)

    print(f"{N}-ECSim:", lle_euclid_similarity, lle_cosine_similarity)
```

```python
import numpy as np
import scipy.linalg as lg
from sklearn.preprocessing import normalize

def train_model(graph, dim):
    A = nx.to_numpy_array(graph, nodelist=sorted(graph.nodes()), weight='weight')
    normalize(A, norm='l1', axis=1, copy=False)
    I_n = np.eye(graph.number_of_nodes())
    #I_min_A = I_n - A
    I_min_A = np.dot((I_n - A).T, (I_n - A))
    u, s, vt = lg.svd(I_min_A, full_matrices=False)
    idx = np.argsort(s) 
    #print(s[idx])
    vt = vt[:, idx]
    vt=vt[:(dim+1)]
    embedding = vt.T[:, 1:]
    #print(embedding[:3])
    return embedding
```

```python
K = 14
for D in range(10,100,5):
    lle_embd = train_model(G, D)

    lle_vec = [Vectors.dense(row) for row in lle_embd]
    lle_prep = SparkSession.builder.getOrCreate().createDataFrame([(vector,) for vector in lle_vec], ["embd"])

    # from pyspark.ml.clustering import KMeans

    # Create and fit the KMeans model
    euclid_kmeans = KMeans(k=K, featuresCol="embd")
    cosine_kmeans = KMeans(k=K, featuresCol="embd", distanceMeasure="cosine")
    lle_euclid_model = euclid_kmeans.fit(lle_prep)
    lle_cosine_model = cosine_kmeans.fit(lle_prep)

    # Add the cluster assignment to the DataFrame
    lle_euclid = lle_euclid_model.transform(lle_prep)
    lle_cosine = lle_cosine_model.transform(lle_prep)

    # hope_euclid.show(truncate=False) # 这条语句可以查看

    # Extract the cluster assignment and convert it to a list
    lle_euclid_membership = lle_euclid.select("prediction").rdd.flatMap(lambda x: x).collect()
    lle_cosine_membership = lle_cosine.select("prediction").rdd.flatMap(lambda x: x).collect()

    # 计算
    # 内在的社群结构 intrinsic_membership 和
    # 嵌入聚类算法
    # 的 NMI 以及 Element centric similarity
    # from sklearn.metrics import normalized_mutual_info_score

    # Calculate the NMI
    lle_euclid_nmi = normalized_mutual_info_score(list(lle_euclid_membership), list(intrinsic_membership.values()))
    lle_cosine_nmi = normalized_mutual_info_score(list(lle_cosine_membership), list(intrinsic_membership.values()))

    print(f"{D}-NMIs:", lle_euclid_nmi, lle_cosine_nmi)

    lle_euclid_clustering = Clustering(elm2clu_dict={i: [lle_euclid_membership[i]] for i in range(len(lle_euclid_membership))})
    lle_cosine_clustering = Clustering(elm2clu_dict={i: [lle_cosine_membership[i]] for i in range(len(lle_cosine_membership))})

    lle_euclid_similarity = sim.element_sim(intrinsic_clustering, lle_euclid_clustering, alpha=0.9)
    lle_cosine_similarity = sim.element_sim(intrinsic_clustering, lle_cosine_clustering, alpha=0.9)

    print(f"{D}-ECSim:", lle_euclid_similarity, lle_cosine_similarity)
```

```python
###############################
### Correct Graph LLE
###############################
import numpy as np
import networkx as nx
from sklearn.preprocessing import normalize
import scipy.linalg as lg

def lle(graph, dim):
    A = nx.to_numpy_array(graph, nodelist=sorted(graph.nodes()), weight='weight')
    normalize(A, norm='l1', axis=1, copy=False)
    I_n = np.eye(graph.number_of_nodes())
    I_min_A = np.dot((I_n - A).T, (I_n - A))
    w, v = lg.eig(I_min_A)
    idx = np.argsort(w.real)
    v = v[:, idx]
    embedding = v[:, 1:(dim+1)]
    return embedding
```

```python
%%time
K = 14
for D in range(10,20):
    lle_embd = lle(G, D)
    lle_vec = [Vectors.dense(row) for row in lle_embd]
    lle_prep = SparkSession.builder.getOrCreate().createDataFrame([(vector,) for vector in lle_vec], ["embd"])

    # from pyspark.ml.clustering import KMeans

    # Create and fit the KMeans model
    euclid_kmeans = KMeans(k=K, featuresCol="embd")
    cosine_kmeans = KMeans(k=K, featuresCol="embd", distanceMeasure="cosine")
    lle_euclid_model = euclid_kmeans.fit(lle_prep)
    lle_cosine_model = cosine_kmeans.fit(lle_prep)

    # Add the cluster assignment to the DataFrame
    lle_euclid = lle_euclid_model.transform(lle_prep)
    lle_cosine = lle_cosine_model.transform(lle_prep)

    # hope_euclid.show(truncate=False) # 这条语句可以查看

    # Extract the cluster assignment and convert it to a list
    lle_euclid_membership = lle_euclid.select("prediction").rdd.flatMap(lambda x: x).collect()
    lle_cosine_membership = lle_cosine.select("prediction").rdd.flatMap(lambda x: x).collect()

    # 计算
    # 内在的社群结构 intrinsic_membership 和
    # 嵌入聚类算法
    # 的 NMI 以及 Element centric similarity
    # from sklearn.metrics import normalized_mutual_info_score

    # Calculate the NMI
    lle_euclid_nmi = normalized_mutual_info_score(list(lle_euclid_membership), list(intrinsic_membership.values()))
    lle_cosine_nmi = normalized_mutual_info_score(list(lle_cosine_membership), list(intrinsic_membership.values()))

    print(f"{D}-NMIs:", lle_euclid_nmi, lle_cosine_nmi)

    lle_euclid_clustering = Clustering(elm2clu_dict={i: [lle_euclid_membership[i]] for i in range(len(lle_euclid_membership))})
    lle_cosine_clustering = Clustering(elm2clu_dict={i: [lle_cosine_membership[i]] for i in range(len(lle_cosine_membership))})

    lle_euclid_similarity = sim.element_sim(intrinsic_clustering, lle_euclid_clustering, alpha=0.9)
    lle_cosine_similarity = sim.element_sim(intrinsic_clustering, lle_cosine_clustering, alpha=0.9)

    print(f"{D}-ECSim:", lle_euclid_similarity, lle_cosine_similarity,"\n")
    
```

```python
import numpy as np
import scipy.sparse as sp
import networkx as nx
from sklearn.preprocessing import normalize
import scipy.sparse.linalg as lg

def lle2(graph, dim):
    A = nx.to_scipy_sparse_matrix(graph, nodelist=sorted(graph.nodes()), weight='weight')
    normalize(A, norm='l1', axis=1, copy=False)
    I_n = sp.eye(graph.number_of_nodes())
    I_min_A = np.dot((I_n - A).T, (I_n - A))
    w, v = lg.eigs(I_min_A)
    idx = np.argsort(np.abs(w))
    v = v[:, idx]
    embedding = v[:, 1:(dim+1)]
    return embedding
```
