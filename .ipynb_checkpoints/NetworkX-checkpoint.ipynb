```python
### 这个文件演示如何用存储在文件中的边集构造 Networkx 图对象
###
### 从文件中读入内容并转化成 边列表
with open('NetworkX.edgelist', 'r') as file:
    lines = file.readlines()

### Process the lines and create a list of number pairs
edge_list = []
for line in lines:
    pair = tuple(map(int, line.strip().split()))
    edge_list.append(pair)

### Print the list of number pairs
print(edge_list)
```

```python
### 导入 networkx 包
import networkx as nx

### 新建一个图 
G = nx.Graph()

### 向图添加边
G.add_edges_from(edge_list)

### 画图
nx.draw(G, with_labels=True)
```

当使用NetworkX创建的图G（使用networkx.Graph()或其他图创建函数）具有各种属性，这些属性提供有关图结构和特性的信息。下面是NetworkX图常用的一些属性：

    G.nodes：类似列表的对象，包含图中的所有节点。
    G.edges：类似列表的对象，包含图中的所有边。
    G.number_of_nodes()：返回图中节点的数量。
    G.number_of_edges()：返回图中边的数量。
    G.degree：类似字典的对象，其中键是节点，值是节点的度（与节点相连的边数）。
    G.adj：类似字典的对象，其中键是节点，值是包含邻居节点及其边属性的字典。
    G.graph：类似字典的对象，存储图级别的属性。
    G.nodes[node]['attribute']：使用节点作为键访问节点属性。
    G.edges[u, v]['attribute']：使用节点对作为键访问边属性。

这只是NetworkX图中可用属性的一些示例。具体的属性及其值可能会因创建和操作图的方式而有所不同。您可以查阅NetworkX文档，了解更多关于图对象的可用属性和方法的详细信息。
