```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = np.loadtxt("MNMF0.5_0.txt")  # Replace this with your actual data

column_labels = [16, 32,64, 128, 256]
index_labels = [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]

# Create a DataFrame using Seaborn heatmap
heatmap_df = pd.DataFrame(data, index=index_labels, columns=column_labels)

# Create a heatmap using Seaborn
sns.heatmap(heatmap_df, cmap='viridis', annot=True, fmt=".3f", linewidths=.5)

# Customize labels and display the heatmap
plt.title("Heatmap, Euclidean NMI")
plt.xlabel('Embedding Dimension')
plt.ylabel('Network Size')
plt.show()
```
