```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

column_labels = [16, 32, 64, 128, 256]
index_labels = [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]

embedding_methods = {
    1: "HOPE",
    2: "LE",
    3: "LLE",
    4: "DeepWalk",
    5: "MNMF",
    6: "LINE",
    7: "node2vec"
}

Kmeans_List = {
    0:"Euclidean NMI",
    1: "Spherical NMI",
    2: "Euclidean ECS",
    3: "Spherical ECS",
}

overall_min = 0
overall_max = 1
k=3
        
for e in [1,2,3,4,5,6]:#7]:
    #for k in range(0,4):
    data = np.loadtxt(f"{e}_{k}.mean")  # Replace this with your actual data

    # Create a DataFrame using Seaborn heatmap
    heatmap_df = pd.DataFrame(data, index=index_labels, columns=column_labels)

    # Create a heatmap using Seaborn
    sns.heatmap(heatmap_df, cmap='viridis', annot=True, fmt=".3f", linewidths=.5, vmin=overall_min, vmax=overall_max)

    # Customize labels and display the heatmap
    plt.title(f"Heatmap, {embedding_methods[e]}, {Kmeans_List[k]}")
    plt.xlabel('Embedding Dimension')
    plt.ylabel('Network Size')
    plt.savefig(f"HM001_{e}_{k}.png")
    plt.show()
```
