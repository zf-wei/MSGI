```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

column_labels = [16, 32, 64, 128, 256]
index_labels = [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]

e = 7

# Generate random data for heatmaps
data1 = np.loadtxt(f"HeatMap_0.01/{e}_{3}.mean")
data2 = np.loadtxt(f"HeatMap_0.5/{e}_{3}.mean")

# Determine the common color scale limits
vmin = 0
vmax = 1

# Set the style for Seaborn
sns.set()

# Create subplots with a shared y-axis and x-axis
fig, axs = plt.subplots(1, 2, figsize=(10, 5), sharey=True, sharex=True, constrained_layout=True)  # 1 row, 2 columns, share y-axis

# Plot the first heatmap in the first subplot using Seaborn
heatmap1 = sns.heatmap(data1, cmap='viridis', ax=axs[0], annot=True, fmt=".2f", cbar=False, vmin=0, vmax=1,
                       xticklabels=column_labels, yticklabels=index_labels)
axs[0].set_title('(a) μ=0.01')
axs[0].set_xlabel('')  # Remove x-axis label
axs[0].set_ylabel('Network Size')

# Plot the second heatmap in the second subplot using Seaborn
heatmap2 = sns.heatmap(data2, cmap='viridis', ax=axs[1], annot=True, fmt=".2f", cbar=False, vmin=0, vmax=1,
                       xticklabels=column_labels, yticklabels=index_labels)
axs[1].set_title('(b) μ=0.5')
axs[1].set_xlabel('')  # Remove x-axis label
axs[1].set_ylabel('')


# Create a common color bar for both heatmaps
cbar_ax = fig.add_axes([1.02, 0.15, 0.02, 0.7])  # [left, bottom, width, height]
cbar = fig.colorbar(heatmap2.collections[0], cax=cbar_ax)

# Set color bar label
cbar.set_label('ECS')

# Set a common x-axis label for both subplots
fig.text(0.52, -0.02, 'Embedding Dimension', ha='center', va='center', fontsize=12)

plt.savefig("test",bbox_inches='tight')

# Show the plot
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

column_labels = [16, 32, 64, 128, 256]
index_labels = [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000]

E = [1,2,3,4,5,7]

for e in E:
    # Generate random data for heatmaps
    data1 = np.loadtxt(f"HeatMap_0.01/{e}_{3}.mean")
    data2 = np.loadtxt(f"HeatMap_0.5/{e}_{3}.mean")

    # Determine the common color scale limits
    vmin = 0
    vmax = 1

    # Set the style for Seaborn
    sns.set()

    # Create subplots with a shared y-axis and x-axis
    fig, axs = plt.subplots(1, 2, figsize=(10, 5), sharey=True, sharex=True, constrained_layout=True)  # 1 row, 2 columns, share y-axis

    # Plot the first heatmap in the first subplot using Seaborn
    heatmap1 = sns.heatmap(data1, cmap="YlOrRd", ax=axs[0], annot=True, fmt=".2f", cbar=False, vmin=0, vmax=1,
                           xticklabels=column_labels, yticklabels=index_labels)
    axs[0].set_title('(a) μ=0.01')
    axs[0].set_xlabel('')  # Remove x-axis label
    axs[0].set_ylabel('Network Size')
    heatmap1.set_yticklabels(heatmap1.get_yticklabels(), rotation=0)

    # Plot the second heatmap in the second subplot using Seaborn
    heatmap2 = sns.heatmap(data2, cmap="YlOrRd", ax=axs[1], annot=True, fmt=".2f", cbar=False, vmin=0, vmax=1,
                           xticklabels=column_labels, yticklabels=index_labels)
    axs[1].set_title('(b) μ=0.5')
    axs[1].set_xlabel('')  # Remove x-axis label
    axs[1].set_ylabel('')


    # Create a common color bar for both heatmaps
    cbar_ax = fig.add_axes([1.02, 0.15, 0.02, 0.7])  # [left, bottom, width, height]
    cbar = fig.colorbar(heatmap2.collections[0], cax=cbar_ax)

    # Set color bar label
    cbar.set_label('ECS')

    # Set a common x-axis label for both subplots
    fig.text(0.52, -0.03, 'Embedding Dimension', ha='center', va='center', fontsize=12)

    plt.savefig(f"MP_{e}.png",bbox_inches='tight')
    
    print(e)
    # Show the plot
    plt.show()
```
