```python
### 这个文件尝试了 Louvain 聚类算法 并 绘制了子图
###
from community import community_louvain
import networkx as nx
import matplotlib

G=nx.karate_club_graph()
```

```python
partition = community_louvain.best_partition(G)
print(partition)
```

```python
pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G,
                      pos,
                      partition.keys(),
                      node_size=40,
                      cmap=matplotlib.colormaps.get_cmap('gist_rainbow'),
                      node_color=list(partition.values())
                      )
nx.draw_networkx_edges(G, pos, alpha=0.5)
#plt.show()
```

```python
### Get the nodes of interest for the induced subgraph
nodes_of_interest = [node for node in G if partition[node] == 0]

### Create the induced subgraph
subgraph = G.subgraph(nodes_of_interest)

### Draw the nodes with their respective community colors
# nodes = 
nx.draw_networkx_nodes(subgraph, pos, node_size=40, cmap=matplotlib.colormaps.get_cmap('gist_rainbow'),
                               node_color=list(partition[node] for node in nodes_of_interest))

### Draw the edges
nx.draw_networkx_edges(subgraph, pos, alpha=0.5)

# Show the graph
#plt.show()
```
